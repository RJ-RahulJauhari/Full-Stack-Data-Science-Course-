Q.1 Can you explain the logic and working of the Tower of Hanoi algorithm by writing a Java program?
How does the recursion work, and how are the movements of disks between rods accomplished?



<p> The problem involves three rods and a number of disks of different sizes that can slide onto any rod. The puzzle starts with the disks in a neat stack in ascending order of size on one rod, the smallest at the top, and the goal is to move the entire stack to another rod, obeying the following simple rules:</p>

<p>Only one disk can be moved at a time.
Each move consists of taking the upper disk from one of the stacks and placing it on top of another stack or on an empty rod.
No disk may be placed on top of a smaller disk.</p>

<p>The base case is when n is 0, which means there are no disks to move, and the function returns without performing any moves. Otherwise, the function makes two recursive calls:</p>
<ul>
    <li>Move (n-1) disks from the source rod to the auxiliary rod using the target rod as an auxiliary peg.</li>
    <li>Move the nth disk from the source rod to the target rod.</li>
    <li>Move (n-1) disks from the auxiliary rod to the target rod using the source rod as an auxiliary peg.</li>
    <li>This process continues recursively until the base case is reached, and the entire Tower of Hanoi problem is solved by breaking it down into smaller subproblems and solving them in a systematic manner.</li>
</ul>nner.isk.

```python
def tower_of_hanoi(n, source, target, auxiliary):
    if n > 0:
        tower_of_hanoi(n-1, source, auxiliary, target)
        move_disk(source, target)
        tower_of_hanoi(n-1, auxiliary, target, source)

def move_disk(source, target):
    print(f"Move disk from {source} to {target}")

tower_of_hanoi(3, 'A', 'C', 'B')
```

Q.2 Given two strings word1 and word2, return the minimum number of operations required to convert word1
to word2on (insert 'u').

```python
def editDistanceHelper(i, j, str1, str2):
 
    if i == 0:
        return j
 
    if j == 0:
        return i
 
    ans = 1 + min(
        {
            editDistanceHelper(i, j - 1, str1, str2),  # Insert
            editDistanceHelper(i - 1, j, str1, str2),  # Remove
            editDistanceHelper(i - 1, j - 1, str1, str2),  # Replace
        }
    )
    if str1[i - 1] == str2[j - 1]:
        ans = min(ans, editDistanceHelper(i - 1, j - 1, str1, str2))
    return ans
 
 
def editDistance(str1, str2):
 
    n = len(str1)
    m = len(str2)
    ans = editDistanceHelper(n, m, str1, str2)
    return ans

editDistance('horse','ros')
```

Q. 3 Print the max value of the array [ 13, 1, -3, 22, 5].

```python
def maxValueOfArray(arr):
    m = arr[0]
    for element in arr:
        if element > m:
            m = element
    return m

def maxValueOfArrayRecursive(arr,m = 0,pointer = 0):
    if pointer == len(arr):
        return m
    else:
        if arr[pointer] > m:
            return maxValueOfArrayRecursive(arr,arr[pointer],pointer+1)
        else:
            return maxValueOfArrayRecursive(arr,m,pointer+1)
    

arr = [13,1,-3,22,5]
# maxValueOfArray(arr)
maxValueOfArrayRecursive(arr)
```

Q.4 Find the sum of the values of the array [92, 23, 15, -20, 10].

```python
def sumOfAllValues(arr):
    s = 0
    for element in arr:
        s += element
    return s

def sumOfAllValuesRecursive(arr,s = 0,pointer = 0):
    if pointer == len(arr):
        return s
    else:
        return sumOfAllValuesRecursive(arr,s + arr[pointer], pointer+1)

# sumOfAllValues([92,23,15,-20,10])
sumOfAllValuesRecursive([92,23,15,-20,10])
```

Q.5 Given a number n. Print if it is an armstrong number or not.An armstrong number is a number if the sum
of every digit in that number raised to the power of total digits in that number is equal to the number.

```python
def extractDigits(num):
    digits = []
    n = num
    while n != 0:
        digit = n % 10
        digits.insert(0,digit)
        n = n // 10
    return digits
    
def isArmstrong(num):
    digits = extractDigits(num)
    exp = len(digits)
    s = 0

    for n in digits:
        s += n**exp
    if s == num:
        return True
    else:
        return False

def isArmstrongRecusive(num,digits = [],s = 0, pointer = 0):
    if pointer == len(digits) and len(digits) != 0:
        return s == num
    if len(digits) == 0:
        return isArmstrongRecusive(num,digits = extractDigits(num),s = 0,pointer = 0)
    else:
        return isArmstrongRecusive(num,digits,s + digits[pointer]**len(digits),pointer + 1)

isArmstrong(153)

isArmstrongRecusive(548834)
```
